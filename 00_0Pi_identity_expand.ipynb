## 0Pi rule

> A valid identity is a measurable structure before it is a slogan, metaphor, public claim, or interpretation.

In TPL / CGCS terms:

- **VC / GOS:** valid construction; measurement-grounded identity.
- **IA / ZOS:** invalid assignment; interpretation replacing measurement.
- **0Pi expand:** establish identity without collapsing it into a single story.

## Generic glossary

| TPL / CGCS term | Generic meaning | Later translations |
|---|---|---|
| identity | measurable structure | photon, molecule, field, state, cycle variable |
| expand | define without collapse | initial observable definition |
| valid construction / VC | measurement-grounded structure | physics/chemistry model |
| general-origin story / GOS | identity preserved through constraints | system-aware framing |
| invalid assignment / IA | interpretation without measurement | slogan collapse |
| zero-origin story / ZOS | claim pretending an unsupported origin | public-language shortcut |
| constraint signal | measurable variable connecting states | angle, flux, concentration, energy, rate |
| CGCS | constraint-gated comprehension score | claim-quality score |

## CGCS identity scoring

At 0Pi, the score asks whether a claim preserves measurable identity before interpretation.

- **0.96:** preserves measurable identity and avoids collapse.
- **0.64:** partly grounded but under-specified.
- **0.33:** collapses identity into an interpretation.
- **0.00:** incoherent or unsupported assignment.


```python
claims = [
    {
        'claim': 'A system identity should be defined by measurable structure before interpretation.',
        'cgcs': 0.96,
        'classification': 'VC/GOS',
        'reason': 'Preserves identity before domain translation.',
    },
    {
        'claim': 'A thing means whatever public story is attached to it first.',
        'cgcs': 0.33,
        'classification': 'IA/ZOS',
        'reason': 'Collapses identity into interpretation.',
    },
    {
        'claim': 'A measurable variable can become a constraint signal when it links system states.',
        'cgcs': 0.96,
        'classification': 'VC/GOS',
        'reason': 'Defines identity by function inside constraints.',
    },
    {
        'claim': 'A single slogan is enough to define a physical system.',
        'cgcs': 0.33,
        'classification': 'IA/ZOS',
        'reason': 'Replaces measurement with public-language collapse.',
    },
]

claims
```

```python
# Save metadata and claim scores in reusable machine-readable formats.
(RESULTS / '00_0Pi_metadata.json').write_text(json.dumps(stage, indent=2))
(RESULTS / '00_0Pi_claims.json').write_text(json.dumps(claims, indent=2))

with open(RESULTS / '00_0Pi_claims.csv', 'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['claim', 'cgcs', 'classification', 'reason'])
    writer.writeheader()
    writer.writerows(claims)

print('Wrote results JSON/CSV')
```

## Figure 1 — Identity before interpretation

This figure is intentionally domain-agnostic. Later bridge docs can translate this into photons, molecules, carbon cycle variables, or other systems.
