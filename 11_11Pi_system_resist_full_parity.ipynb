## 11Pi rule

> A valid system preserves coordination under constraint pressure.

In TPL / CGCS terms:

- **VC / GOS:** valid construction; coupling and feedback remain bounded and coherent under constraint.
- **IA / ZOS:** invalid assignment; untracked coupling or feedback is treated as valid system behavior.
- **11Pi resist:** preserve system coordination through a constraint gate.
