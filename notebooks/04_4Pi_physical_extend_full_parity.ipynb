## 4Pi rule

> A valid physical interaction remains consistent under variable change.

In TPL / CGCS terms:

- **VC / GOS:** valid construction; physical response remains consistent across measurable variables.
- **IA / ZOS:** invalid assignment; variable change collapses interaction into interpretation.
- **4Pi extend:** probe physical interaction across parameter variation.
