## 7Pi rule

> A valid cycle remains consistent across rates, reservoirs, and pathways.

In TPL / CGCS terms:

- **VC / GOS:** valid construction; cycle exchange remains measurable across rates, reservoirs, and pathways.
- **IA / ZOS:** invalid assignment; pathways are ignored or reservoir shifts are treated as arbitrary.
- **7Pi extend:** extend repeatable cycle exchange through structured variation.
