```python
from pathlib import Path
import json, csv, zipfile, matplotlib.pyplot as plt

ROOT=Path(".")
DOCS=ROOT/"docs"
RESULTS=ROOT/"results"
FIGURES=ROOT/"figures"
MATH=ROOT/"math"

for f in [DOCS,RESULTS,FIGURES,MATH]:
    f.mkdir(exist_ok=True)

print("Initialized folders")
```

```python
stage={
"notebook":"07_7Pi_cycle_extend",
"pi_stage":"7Pi",
"triplet_position":"extend",
"track":"cycle",
"core_rule":"Extend cycle exchange across rates, reservoirs, and pathways.",
"template_role":"Full parity cycle extension notebook",
"previous_stage":"06_6Pi_cycle_expand",
"next_stage":"08_8Pi_cycle_resist",
"cgcs_target":0.96,
"outputs":["docs","results","figures","math"]
}

(RESULTS/"07_7Pi_metadata.json").write_text(json.dumps(stage,indent=2))
stage
```

```python
claims=[
{"claim":"A valid cycle remains consistent across rates, reservoirs, and pathways.","cgcs":0.96,"classification":"VC/GOS","reason":"cycle continuity"},
{"claim":"Changing rates or reservoirs can redefine a cycle arbitrarily.","cgcs":0.33,"classification":"IA/ZOS","reason":"breaks cycle consistency"},
{"claim":"Cycle extension requires measurable transfer among states.","cgcs":0.96,"classification":"VC/GOS","reason":"observable exchange"},
{"claim":"A cycle can be explained without tracking pathways.","cgcs":0.33,"classification":"IA/ZOS","reason":"removes structure"}
]

(RESULTS/"07_7Pi_claims.json").write_text(json.dumps(claims,indent=2))

import csv
with open(RESULTS/"07_7Pi_claims.csv","w",newline="") as f:
    w=csv.DictWriter(f,fieldnames=["claim","cgcs","classification","reason"])
    w.writeheader(); w.writerows(claims)

claims
```

```python
fig,ax=plt.subplots(figsize=(6,6))
pts=[(0,1),( -1,-1),(1,-1)]
labels=["reservoir A","reservoir B","reservoir C"]

for (x,y),l in zip(pts,labels):
    ax.scatter([x],[y],s=1000)
    ax.text(x,y,l,ha="center",va="center")

ax.annotate("",xy=(-0.8,-0.8),xytext=(0,0.8),arrowprops=dict(arrowstyle="->",lw=2))
ax.annotate("",xy=(0.8,-0.8),xytext=(-0.8,-0.8),arrowprops=dict(arrowstyle="->",lw=2))
ax.annotate("",xy=(0,0.8),xytext=(0.8,-0.8),arrowprops=dict(arrowstyle="->",lw=2))

ax.text(0,1.5,"7Pi extend cycle
rates · reservoirs · pathways",ha="center")

ax.axis("off")
p=FIGURES/"07_7Pi_cycle_extend.png"
plt.savefig(p)
print(p)
```

```python
fig,ax=plt.subplots(figsize=(8,4))
ax.plot([0.5,0.5],[0,1])

ax.text(0.25,0.7,"IA / ZOS",ha="center")
ax.text(0.25,0.4,"pathway ignored",ha="center")
ax.text(0.25,0.2,"cycle collapse",ha="center")

ax.text(0.75,0.7,"VC / GOS",ha="center")
ax.text(0.75,0.4,"tracked pathways",ha="center")
ax.text(0.75,0.2,"extended exchange",ha="center")

ax.set_title("7Pi CGCS gate")
ax.axis("off")

p=FIGURES/"07_7Pi_vc_gos_vs_ia_zos.png"
plt.savefig(p)
print(p)
```
