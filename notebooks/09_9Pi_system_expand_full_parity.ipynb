## 9Pi rule

> A stable cycle becomes a system when multiple constrained exchanges coordinate.

In TPL / CGCS terms:

- **VC / GOS:** valid construction; system behavior is grounded in coordinated cycle exchange.
- **IA / ZOS:** invalid assignment; one isolated cycle is treated as a full system.
- **9Pi expand:** expand stable cycle behavior into coupled system structure.
