2Pi resists identity collapse by preserving measurable structure across constraint.


## 2Pi rule

> A valid identity remains invariant under constraint.

- **VC / GOS:** identity preserved across constraint
- **IA / ZOS:** identity collapses under constraint
- **2Pi resist:** enforce preservation after interaction


```python
from pathlib import Path
import json, csv, zipfile
import matplotlib.pyplot as plt

ROOT = Path(".")
DOCS = ROOT / "docs"
RESULTS = ROOT / "results"
FIGURES = ROOT / "figures"
MATH = ROOT / "math"

for f in [DOCS, RESULTS, FIGURES, MATH]:
    f.mkdir(exist_ok=True)

print("Initialized:", DOCS, RESULTS, FIGURES, MATH)
```

## Metadata

```python
stage = {
    "notebook": "02_2Pi_identity_resist",
    "pi_stage": "2Pi",
    "triplet_position": "resist",
    "track": "identity",
    "core_rule": "Preserve identity across constraint.",
    "previous_stage": "01_1Pi_identity_extend",
    "next_stage": "03_3Pi_physical_expand",
    "cgcs_target": 0.96,
    "outputs": ["docs","results","figures","math"],
}
stage
```

## Claims

```python
claims = [
    {"claim":"A valid identity is preserved across constraint pressure.","cgcs":0.96,"classification":"VC/GOS","reason":"Identity stable after interaction"},
    {"claim":"Constraint erases identity.","cgcs":0.33,"classification":"IA/ZOS","reason":"Collapse confusion"},
    {"claim":"Preservation requires measurable continuity.","cgcs":0.96,"classification":"VC/GOS","reason":"Continuity enforced"},
    {"claim":"Interpretation can replace identity after constraint.","cgcs":0.33,"classification":"IA/ZOS","reason":"Interpretation collapse"}
]
claims
```

```python
(RESULTS/"02_2Pi_metadata.json").write_text(json.dumps(stage,indent=2))
(RESULTS/"02_2Pi_claims.json").write_text(json.dumps(claims,indent=2))

with open(RESULTS/"02_2Pi_claims.csv","w",newline="") as f:
    w=csv.DictWriter(f,fieldnames=["claim","cgcs","classification","reason"])
    w.writeheader()
    w.writerows(claims)

print("Saved results")
```

## Figure 1

```python
fig,ax=plt.subplots(figsize=(9,3))
x=[0,1,2]
labels=["defined identity","interaction","preserved identity"]
ax.scatter(x,[0,0,0],s=900)

for xi,l in zip(x,labels):
    ax.text(xi,-0.4,l,ha="center")

ax.annotate("",xy=(0.8,0),xytext=(0.2,0),arrowprops=dict(arrowstyle="->",lw=2))
ax.annotate("",xy=(1.8,0),xytext=(1.2,0),arrowprops=dict(arrowstyle="->",lw=2))

ax.text(0,0.3,"0Pi",ha="center")
ax.text(1,0.3,"1Pi",ha="center")
ax.text(2,0.3,"2Pi",ha="center")

ax.set_title("2Pi identity resist: preservation across constraint")
ax.axis("off")

p=FIGURES/"02_2Pi_identity_resist.png"
plt.savefig(p,dpi=200,bbox_inches="tight")
plt.show()
print("Saved",p)
```
