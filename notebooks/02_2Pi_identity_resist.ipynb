## 2Pi rule

> A valid identity is preserved across constraint pressure.

In TPL / CGCS terms:

- **VC / GOS:** valid construction; identity stabilized across constraint.
- **IA / ZOS:** invalid assignment; constraint collapses identity into interpretation.
- **2Pi resist:** preserve identity after definition and interaction testing.
