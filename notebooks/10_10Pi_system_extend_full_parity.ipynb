## 10Pi rule

> A valid system remains coherent across scale, coupling strength, and feedback.

In TPL / CGCS terms:

- **VC / GOS:** valid construction; system coherence is preserved across measurable variation.
- **IA / ZOS:** invalid assignment; feedback or coupling is ignored while claiming system behavior.
- **10Pi extend:** extend coordinated system structure through scale, coupling, and feedback variation.

## Figure 1 — System extension
