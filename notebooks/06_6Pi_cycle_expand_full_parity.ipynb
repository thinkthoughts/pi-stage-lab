## 6Pi rule

> A stable physical signal becomes a cycle variable when it participates in repeatable exchange.

In TPL / CGCS terms:

- **VC / GOS:** valid construction; cycle behavior is grounded in repeatable exchange.
- **IA / ZOS:** invalid assignment; a single event is treated as a cycle.
- **6Pi expand:** expand stable physical signal into repeatable cycle exchange.
