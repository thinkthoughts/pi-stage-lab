## 5Pi rule

> A valid physical signal remains stable across constraint pressure.

In TPL / CGCS terms:

- **VC / GOS:** valid construction; signal is preserved across constraint pressure.
- **IA / ZOS:** invalid assignment; unstable measurements are treated as stable behavior.
- **5Pi resist:** preserve physical signal through a constraint gate.
