## 13Pi rule

> A valid rate remains meaningful across intervals, baselines, and comparisons.

- **VC / GOS:** rate comparison is grounded in explicit interval and baseline.
- **IA / ZOS:** rate is compared without measurement context.
- **13Pi extend:** extend measured rate into comparable rate context.

## Generic glossary

| TPL / CGCS term | Generic meaning | Later translations |
|---|---|---|
| rate | measured change per interval | concentration change, flux rate, warming rate |
| extend | compare rate across context | intervals, baselines, comparisons |
| interval | counted step or measurement span | Δn, sample step, observation interval |
| baseline | reference quantity or condition | reference concentration, reference flux |
| comparison | relation between rates or rate contexts | relative rate, normalized rate |
| comparable rate | rate normalized to explicit context | rate ratio, baseline-normalized change |
| context collapse | rate claim losing interval or baseline | misleading comparison |
| valid construction / VC | rate comparison grounded in context | physics/chemistry model |
| invalid assignment / IA | comparison without baseline or interval | slogan collapse |
| CGCS | constraint-gated comprehension score | claim-quality score |
