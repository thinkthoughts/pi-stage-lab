## 16Pi rule

> A valid translation remains consistent across contexts, audiences, and representations.

- **VC / GOS:** mapped meaning remains preserved across expression contexts.
- **IA / ZOS:** audience or representation changes redefine measured meaning.
- **16Pi extend:** extend mapped language through context, audience, and representation boundaries.

- **0.00:** incoherent or unsupported assignment.
