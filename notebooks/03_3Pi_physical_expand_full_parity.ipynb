## 3Pi rule

> A preserved identity becomes physically meaningful through measurable interaction.

In TPL / CGCS terms:

- **VC / GOS:** valid construction; physical meaning is grounded in measurable interaction.
- **IA / ZOS:** invalid assignment; interpretation is treated as physical behavior without measurement.
- **3Pi expand:** expand preserved identity into observable physical signal.
