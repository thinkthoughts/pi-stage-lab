## 15Pi rule

> A valid translation preserves measurable rate structure in language.

- **VC / GOS:** measured structure is mapped into language without losing interval, baseline, or comparison context.
- **IA / ZOS:** interpretation-only language replaces measurement structure.
- **15Pi expand:** expand stable rate structure into mapped language expression.
