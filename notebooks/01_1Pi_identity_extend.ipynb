## 1Pi rule

> A valid identity must remain consistent through interaction.

In TPL / CGCS terms:

- **VC / GOS:** valid construction; identity persists through interaction.
- **IA / ZOS:** invalid assignment; interaction collapses identity into interpretation.
- **1Pi extend:** test identity under measurement, exchange, perturbation, or transformation.

## CGCS identity scoring

At 1Pi, the score asks whether a claim preserves measurable identity through interaction.

- **0.96:** identity remains consistent under interaction.
- **0.64:** partly stable but under-specified.
- **0.33:** interaction collapses identity into interpretation.
- **0.00:** incoherent or unsupported assignment.


````python
claims = [
    {
        "claim": "A valid identity must remain consistent under interaction.",
        "cgcs": 0.96,
        "classification": "VC/GOS",
        "reason": "Identity persists under transformation.",
    },
    {
        "claim": "Interaction can redefine identity completely without measurement.",
        "cgcs": 0.33,
        "classification": "IA/ZOS",
        "reason": "Breaks identity continuity.",
    },
    {
        "claim": "A measurement should not redefine the identity itself.",
        "cgcs": 0.96,
        "classification": "VC/GOS",
        "reason": "Separates observation from identity.",
    },
    {
        "claim": "A perturbation is enough to replace a system identity with a public interpretation.",
        "cgcs": 0.33,
        "classification": "IA/ZOS",
        "reason": "Replaces measurable interaction with interpretation collapse.",
    },
]

claims
````

## Figure 1 — Identity through interaction

This figure is intentionally domain-agnostic.

1Pi asks whether identity still holds after interaction.


````python
fig, ax = plt.subplots(figsize=(9, 3.4))

x = [0, 1, 2]
labels = ["defined\nidentity", "interaction", "identity\nholds?"]

ax.scatter(x, [0, 0, 0], s=900)

for xi, label in zip(x, labels):
    ax.text(xi, -0.42, label, ha="center", va="top", fontsize=12)

ax.annotate("", xy=(0.78, 0), xytext=(0.22, 0), arrowprops=dict(arrowstyle="->", lw=2.5))
ax.annotate("", xy=(1.78, 0), xytext=(1.22, 0), arrowprops=dict(arrowstyle="->", lw=2.5))

ax.text(0, 0.32, "0Pi define", ha="center", va="bottom", fontsize=10)
ax.text(1, 0.32, "1Pi test", ha="center", va="bottom", fontsize=10)
ax.text(2, 0.32, "identity holds?", ha="center", va="bottom", fontsize=10)

ax.set_title("1Pi identity extend: test through interaction")
ax.set_xlim(-0.45, 2.45)
ax.set_ylim(-0.85, 0.65)
ax.axis("off")

fig_path = FIGURES / "01_1Pi_identity_extend.png"
plt.savefig(fig_path, dpi=200, bbox_inches="tight")
plt.show()

print("Saved", fig_path)
````
