## 12Pi rule

> A stable system becomes rate-aware when change is measured per interval.

- **VC / GOS:** rate is grounded in measured change across an interval.
- **IA / ZOS:** urgency or interpretation is treated as rate without measurement.
- **12Pi expand:** expand stable system behavior into a measured rate signal.
