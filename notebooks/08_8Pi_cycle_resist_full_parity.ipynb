## 8Pi rule

> A valid cycle preserves exchange under constraint pressure.

In TPL / CGCS terms:

- **VC / GOS:** valid construction; cycle exchange remains stable under constraint.
- **IA / ZOS:** invalid assignment; untracked exchange is treated as a valid cycle.
- **8Pi resist:** preserve cycle exchange, pathways, and bounded variation under constraint.

## Figure 1 — Cycle resistance

This figure is intentionally domain-agnostic.

8Pi completes the cycle triplet by preserving exchange under a constraint gate.


````python
fig, ax = plt.subplots(figsize=(7.6, 6.6))

nodes = {
    "reservoir\nA": (0, 1.2),
    "reservoir\nB": (-1.25, -0.75),
    "reservoir\nC": (1.25, -0.75),
}

# outer constraint gate
gate = Circle((0, 0.0), radius=1.75, fill=False, linewidth=2.2)
ax.add_patch(gate)

for label, (x, y) in nodes.items():
    ax.scatter([x], [y], s=1200)
    ax.text(x, y, label, ha="center", va="center", fontsize=12)

# cycle arrows around triangle
ax.annotate("", xy=(-1.02, -0.45), xytext=(-0.12, 0.93), arrowprops=dict(arrowstyle="->", lw=2.4))
ax.annotate("", xy=(1.02, -0.75), xytext=(-1.02, -0.75), arrowprops=dict(arrowstyle="->", lw=2.4))
ax.annotate("", xy=(0.12, 0.93), xytext=(1.02, -0.45), arrowprops=dict(arrowstyle="->", lw=2.4))

ax.text(0, 1.82, "8Pi resist cycle\nconstraint gate", ha="center", va="center", fontsize=13, weight="bold")
ax.text(0, 0.05, "cycle\nstability", ha="center", va="center", fontsize=12)
ax.text(-1.85, 0.1, "preserved\npathways", ha="center", va="center", fontsize=10)
ax.text(1.85, 0.1, "bounded\nvariation", ha="center", va="center", fontsize=10)

ax.set_title("8Pi cycle resist: preserved exchange under constraint")
ax.set_xlim(-2.25, 2.25)
ax.set_ylim(-1.9, 2.15)
ax.axis("off")

fig_path = FIGURES / "08_8Pi_cycle_resist.png"
plt.savefig(fig_path, dpi=200, bbox_inches="tight")
plt.show()

print("Saved", fig_path)
````
